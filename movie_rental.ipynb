```python
import pandas as pd
import numpy as np

from sklearn.model_selection import train_test_split
from sklearn.metrics import mean_squared_error
from sklearn.ensemble import RandomForestRegressor
from sklearn.metrics import mean_squared_error as MSE

# Read data file
rental_df = pd.read_csv('rental_info.csv')
rental_df.head()
```

```python
rental_df['rental_date'] = pd.to_datetime(rental_df['rental_date'])
rental_df['rental_date']
```

```python
rental_df['rental_date'] = pd.to_datetime(rental_df['rental_date'])
rental_df['rental_date']
```

```python
rental_df['return_date'] = pd.to_datetime(rental_df['return_date'])
rental_df['return_date']
```

```python
rental_df['rental_length_days'] = (rental_df['return_date'] - rental_df['rental_date']).dt.days
#rental_df['rental_length_days'] = rental_df['rental_length_days'].dt.days
rental_df.head()
```

```python
rental_df['deleted_scenes'] = np.where(
    rental_df['special_features'].str.contains('Deleted Scenes'), 
    1,  # Value to set if string IS found
    0   # Value to set if string is NOT found
)
rental_df['behind_the_scenes'] = np.where(
    rental_df['special_features'].str.contains('Behind the Scenes'), 
    1,  # Value to set if string IS found
    0   # Value to set if string is NOT found
)
rental_df.head()
```

```python
y = rental_df['rental_length_days']
X = rental_df.drop(['rental_length_days', 
                    'special_features', 
                    'rental_date', 
                    'return_date'], axis=1)
SEED = 9
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.2,
                                                    random_state=SEED)

best_model = RandomForestRegressor(n_estimators=10, random_state=SEED)
best_model.fit(X_train, y_train)
y_pred = best_model.predict(X_test)
best_mse = MSE(y_test, y_pred)
#best_mse
y_pred
```
